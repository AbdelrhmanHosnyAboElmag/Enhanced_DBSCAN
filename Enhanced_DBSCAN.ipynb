```python
from sklearn.datasets import make_moons
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.cluster import DBSCAN
```

```python
#make sperhical dataset
X, _ = make_moons(n_samples=1000)
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
#make non spherical dataset
f, _ = make_blobs(n_samples=1000, centers=3, n_features=2,random_state=0, cluster_std=0.25)
plt.scatter(f[:,0], f[:,1])
plt.show()
```

```python
#apply kmeans to spherical dataset
kmeans = KMeans(n_clusters=2)
results = kmeans.fit_predict(X)
```

```python
#apply kmeans to non-spericahl
kmeansNon = KMeans(n_clusters=3)
resultsNon = kmeansNon.fit_predict(f)
```

```python
#cluster centers of kmeans of sperichal
clusters = kmeans.cluster_centers_
clusters
```

```python
#cluster centers of kmeans of non-sperical
clustersNon = kmeansNon.cluster_centers_
clustersNon
```

```python
#result of kmean in spehrical dataset
plt.scatter(X[:,0], X[:,1], c=results, cmap='coolwarm')
plt.show()
```

```python
#result of kmean in non-spehrical dataset
plt.scatter(f[:,0], f[:,1], c=resultsNon, cmap="coolwarm")
plt.show()
```

```python
#create emty list 2 and 3
list2=[]
list3=[]
list2Non=[]
list3Non=[]
```

```python
#divide clusters to spherical
for i in range (1000):
    if results[i]==0:
        list2.append([X[i][0],X[i][1]])
    else:
        list3.append([X[i][0],X[i][1]])  
#divide clusters to non spherical
for i in range (1000):
    if resultsNon[i]==0:
        list2Non.append([f[i][0],f[i][1]])
    else:
        list3Non.append([f[i][0],f[i][1]])  
```

```python
#add to dataset sperical to use in Dscan
data2 = np.array(list2)
data3 = np.array(list3)
#add to dataset non-sperical to use in Dscan
data2Non = np.array(list2Non)
data3Non = np.array(list3Non)
```

```python
#show dataset sperical
plt.scatter(data2[:,0],data2[:,1])
plt.show()
#show dataset non-sperical
plt.scatter(data2Non[:,0],data2Non[:,1])
plt.show()
```

```python
#show dataset sperical
plt.scatter(data3[:,0],data3[:,1])
plt.show()
#show dataset non-sperical
plt.scatter(data3Non[:,0],data3Non[:,1])
plt.show()
```

```python
#dbscan apply sperical
dbscan = DBSCAN(eps=0.5, min_samples=3)
dbscan.fit(data2)
dbscan.labels_
dbscan1 = DBSCAN(eps=0.5, min_samples=3)
dbscan1.fit(data3)
dbscan1.labels_
#dbscan apply non-sperical
dbscanNon = DBSCAN(eps=0.5, min_samples=3)
dbscanNon.fit(data2Non)
dbscanNon.labels_
dbscan1Non = DBSCAN(eps=0.5, min_samples=3)
dbscan1Non.fit(data3Non)
dbscan1Non.labels_
```

```python
#sperical dataset
plt.scatter(data2[:,0],data2[:,1],c=dbscan.labels_,cmap="coolwarm")
plt.show()
#non sperical dataset
plt.scatter(data2Non[:,0],data2Non[:,1],c=dbscanNon.labels_,cmap="coolwarm")
plt.show()
```

```python
#sperical dataset
plt.scatter(data3[:,0],data3[:,1],c=dbscan1.labels_,cmap="coolwarm")
plt.show()
#non-sperical dataset
plt.scatter(data3Non[:,0],data3Non[:,1],c=dbscan1Non.labels_,cmap="coolwarm")
plt.show()
```
